**Broadcasting** <br>
The term broadcasting refers to the ability of NumPy to treat arrays of different shapes during arithmetic operations. Arithmetic
operations on arrays are usually done on corresponding elements. If two arrays are of exactly the same shape, then these
operations are smoothly performed.
If the dimensions of two arrays are dissimilar, element-to-element operations are not possible. However, operations on arrays
of non-similar shapes is still possible in NumPy, because of the broadcasting capability. The smaller array is broadcast to the
size of the larger array so that they have compatible shapes.


**Rules for Broadcasting**

```python
# If x=m and y=n, operation will take place
a1=np.arange(8).reshape(2,4)
a2=np.arange(8,16).reshape(2,4)

a1+a2
```

```python
# If x=1 and y=n then also operation will take place(Same Dimension)
a5=np. arange(3).reshape(1,3)
a6=np.arange(12).reshape(4,3)
print(a5)
print(a6)
```

```python
a5+a6
```

```python
# If y=1 and x=m then also operation will take place, even if
# they are not of the same dimension
a7=np.arange(4).reshape(4,1)
a8=np.arange (12).reshape(4,3)
print (a7)
print (a8)
```

```python
a7+a8
```

```python
#If x=1 and y!=n then also operation will not take place
a9=np.arange(3).reshape (1,3)
a10=np.arange(16).reshape(4,4)
print(a9)
print (a10)
```

```python
a9+a10
```

```python
# If x=1 and n=1 then y==m, operation to take place
a11=np.arange(3).reshape(1,3)
a12=np.arange(3).reshape(3,1)
print(a11)
print (a12)
```

```python
a11+a12
```

```python
# If x=1 and y=1, then the operation will take place no matter what
a13=np.arange(1,2).reshape(1,1)
a14=np. arange(20).reshape(4,5)
print(a13)
print(a14)
```

```python
a13+a14
```

```python
# If they are of different dimensions
a15=np.arange(4)
a16=np.arange(20).reshape(5,4)
print (a15)
print(a16)
```

```python
a15+a16
```
